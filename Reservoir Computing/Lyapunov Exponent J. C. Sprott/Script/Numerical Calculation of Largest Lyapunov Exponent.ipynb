```python
d0 = 10**-12
```

```python
import numpy as np
from tqdm import tqdm
z = []
for j in tqdm(range(0,1000)):
    dis = []
    for i in range(len(df1)):  
        try:
            d1 = ((xa[i]-xb[i])**2 + (ya[i]-yb[i])**2 + (za[i]-zb[i])**2)**(0.5)
            lg = np.log(d1/d0)
            xb[i] = xa[i+1] + d0*(xb[i+1] - xa[i+1])/d1
            yb[i] = ya[i+1] + d0*(yb[i+1] - ya[i+1])/d1
            zb[i] = za[i+1] + d0*(zb[i+1] - za[i+1])/d1  
            dis.append(lg)
        except:
            pass
    x = np.mean(dis)
    z.append(x)
    print(x)
```
