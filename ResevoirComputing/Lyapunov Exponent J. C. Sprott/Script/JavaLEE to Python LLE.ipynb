```python
#Lorentz_solver
tmax, n = 50, 2829
t = np.linspace(0, 1, 1)
def lorenz(X, t, sigma, beta, rho):
    """The Lorenz equations."""
    u, v, w = X
    up = -sigma*(u - v)
    vp = rho*u - v - u*w
    wp = -beta*w + u*v
    return up, vp, wp
```

```python
def evolveSystem(times):
    for i in range(times):
        evolve()
```

```python
def evolve():
    X1 = odeint(lorenz, ( 0.1 , 1.2, 1.2), t, args=(10.0, 28.0, 8.0/3.0))     
    x1, y1, z1 = X1.T
    return  [np.array(x1), np.array(y1), np.array(z1)]
```

```python
def evolveOther(xOther):
    X2 = odeint(lorenz, xOther, t, args=(10.0, 28.0, 8.0/3.0)) 
    x2, y2, z2 = X2.T
    return [np.array(x2), np.array(y2), np.array(z2)] 
```
